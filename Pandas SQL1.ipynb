## 1. Create an sqlalchemy engine using a sample from the data set

```python

import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',sep=',',header=None,index_col=False)
```

```python
df.head()
```

```python

df.columns =["age","workclass","fnlwgt","education","education_num","matrial_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","income"]
```

```python
df.head()
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String
Base = declarative_base()
```

```python
class mlStudents(Base):
        __tablename__ = 'machineLearning'

        id=Column(Integer, primary_key=True)
        age=Column(Integer)
        workclass=Column(String(250))
        fnlwgt=Column(Integer)
        education = Column(String(250))
        education_num = Column(Integer)
        matrial_status =Column(String(250))
        occupation=Column(String(250))
        relationship=Column(String(250))
        race = Column(String(250))
        sex = Column(String(250))
        capital_gain = Column(Integer)
        capital_loss=Column(Integer)
        hours_per_week=Column(Integer)
        native_country = Column(String(250))
        income = Column(String(250))
```

```python
engine = create_engine('sqlite:///myData.db' , echo=True)
Base.metadata.create_all(engine)
```

```python

df.to_sql('machineLearning', engine,  index=True, index_label='id', if_exists='append')
```
